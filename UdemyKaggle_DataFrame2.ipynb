scikit learnのSimpleImputerを使う

```python
df = pd.read_csv('/content/drive/MyDrive/kaggle_udemy/udemy_course/sample1_without_index.csv', usecols=['Width','Height'])
```

```python
df.head(2)
```

```python
from sklearn.impute import SimpleImputer
```

```python
mena_imputer = SimpleImputer()
```

```python
mena_imputer.fit(df[['Width']])
```

```python
df['Width_Inputed'] = mena_imputer.transform(df[['Width']])
```

```python
df[df['Width'].isnull()]
```

```python
df.head()
```

```python
median_imputer = SimpleImputer(strategy='median')
```

```python
df['Height_Imputed'] = median_imputer.fit_transform(df[['Height']])
```
